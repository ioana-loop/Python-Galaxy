# Observed Brightnesses

```python
#Show descriptive statistics
obs_bright =galaxy.loc[:,'S280MAG':'e.W914FE']
obs_bright.describe()
```

```python
obs_bright2 =galaxy[['S280MAG','W420FE', 'W462FE', 'W485FD','W518FE', 'W571FS', 'W604FE', 'W646FD', 'W696FE', 'W753FE', 'W815FS', 'W856FD', 'W914FD', 'W914FE', 'UFS', 'BFS', 'VFD', 'IFD']]
obs_bright3 =galaxy[['W462FE', 'W485FD','W518FE', 'W571FS', 'W604FE', 'W646FD', 'W696FE', 'W753FE', 'W815FS', 'W856FD', 'W914FD', 'W914FE', 'UFS', 'BFS', 'VFD', 'IFD']]

obs_bright3.describe()
```

```python
#remove data points more tha 2 standard deviation away from the mean
bright_numeric = obs_bright3.select_dtypes([np.number]).columns.values #first, find only columns with numeric data
bright_1 = obs_bright3.loc[:,bright_numeric]

changed = bright_1.copy()
before = bright_1.copy()

total_changed = 0

#for loop for every column
for col in changed.columns:
    #find mean and std of the columns
    col_mean = changed[col].mean()
    col_std = changed[col].std()
    
    #perform outlier test: if abs(X-mean) > 2*std, then replace with mean
    changed.loc[(abs(changed[col] - col_mean)) > (1 * col_std),col] = col_mean
    
    #count number changed in the column. before the subtraction sign counts the number of rows in the column
    #after the subtraction sign use a boolean matrix. Checks to see which rows remained the same. Then adds
    #those rows that remained the same. Expect changes = 0 if no changes
    changes = before[col].count() - (before[col] == changed[col]).sum()
    
    #add changes for each row to the counter for total
    total_changed += changes

print(f'The number of outliers that were replaced with the mean is {total_changed}.')
```
